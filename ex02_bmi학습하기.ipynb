```python
# 그래프를 그리는 함수 정의
def bmi_draw(label, color):
    d = data.loc[label]
    plt.scatter(d['Height'],d['Weight'],c=color,label=label)
```

### 모델링
- 모델 선택 및 하이퍼 파라미터 조정(knn모델 선택)
- 모델 학습
- 모델 예측 및 평가

```python
# 인덱스에 있는 label을 컬럼쪽으로 연결하기(이동하기) : df.reset_index()
# 컬럼명 확인 df.columns
data2 = data.reset_index() # 원본을 유지하고 싶다면 새로운 변수명으로 대입
data2
# data index가 label 로 된 상태가 그대로 유지됨(원본유지)
```

```python
data2
```

```python
# 문제, 답(컬럼 분리)
X = data2[['Height','Weight']]
y = data2['Label']

# 크기확인
print('문제크기: ',X.shape)
print('답크기: ',y.shape)
```

```python
# 행이 총 몇개? 500
# train, test데이터로 분리하는 비율
# 7:3 (7.5:2.5),(8:2)
# X_train, y_train, X_test, y_test
X_train = X[:350] 
X_test = X[350:]

y_train = y[:350]
y_test = y[350:]

print('훈련용 데이터 크기:',X_train.shape,y_train.shape)
print('테스트용 데이터 크기:',X_test.shape,y_test.shape)
# X_train 훈련용 문제, y_train 훈련용 답 데이터
# X_test 테스트용 문제, y_test 테스트용 답 데이터
```

```python
knn_model = KNeighborsClassifier(n_neighbors=10)
```

```python
knn_model.fit(X_train, y_train)
```

```python
# 모델 예측
# knn_model.predict(테스트용 문제)
pre = knn_model.predict(X_test)
pre # 예측 결과를 담은 변수
# 예측이 실제 정답과 항상 같지는 않음, 예측이 틀릴 수 있음(오류가 있을 수 있음)
```

```python
# 모델 평가
# accuracy_score(실제 답, 예측 결과) - 예측이 실제 값과 비교했을 때 몇 개나 잘 맞췄는지 표현하는 점수 척도
# 0~1로 출력, 1에 가까울수록 잘 맞춤
accuracy_score(y_test, pre) # 1.0*100 = 100% 전체 다 맞춘 상황
```
