### H(x)
- 

```python
#  
def h(w, x):
    return w * x + 0
```

### (Cost Function)
-  = 

```python
#  
# data :  
# target : ( )
# weight : 

def cost(data,target,weight):
    # 
    y_pre = h(weight,data)
    #  = (( - )**2).mean()
    error = ((y_pre - target)**2).mean()
    return error
```

```python
#    
cost(data[''],data[''],8)
```

```python
cost(data[''],data[''],7)
```

```python
cost(data[''],data[''],10)
```

```python
#       
cost_list = []
for w in range(5,16): # 5 ~ `15  
    err = cost(data[''],data[''],w)
    cost_list.append(err)
cost_list
```

```python
# (MSE)  
plt.plot(range(5,16),cost_list)
```
